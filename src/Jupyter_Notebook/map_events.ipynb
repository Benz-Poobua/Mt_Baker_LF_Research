```python
import os
import pandas as pd
import matplotlib.pyplot as plt

DATA = os.path.join('..','..','data','Event_Metadata','PNSN_LF_catalog_full_formatted.csv')
```

```python
df = pd.read_csv(DATA, parse_dates=True)
df = df[(df.lat > 0) & (df.lon < 0)]
display(df.evid.value_counts())
```

```python
fig = plt.figure()
gs = fig.add_gridspec(nrows=2,ncols=2)
axes = [fig.add_subplot(gs[i_]) for i_ in range(4)]

axes[0].scatter(df.lon, df.lat, s=df.magnitude)
axes[0].scatter
plt.show()
```
