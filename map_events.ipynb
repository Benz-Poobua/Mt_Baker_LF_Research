```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('./pandafied_le_and_lf_full_catalog.csv', parse_dates=True)
df = df[(df.lat > 0) & (df.lon < 0)]
display(df.min())
display(df.max())
display(df.mean())
```

```python
fig = plt.figure()
gs = fig.add_gridspec(nrows=2,ncols=2)
axes = [fig.add_subplot(gs[i_] for i_ in range(4)]

axes[0][0].scatter(df.lon, df.lat, s=df.magnitude)
plt.show()
```
